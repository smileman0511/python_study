# 마크다운(텍스트 셀)
    - 일반 텍스트 기반의 경량 마크업 언어로 웹 표준이 아닌 언어이다. 마크업이란, 태그 등을 이용하여 문서나 데이터의 구조 등을 기술하며, 텍스트만으로 서식이 있는 문서들을 작성할 때 자주 사용된다. 또한 다른 마크업 언어들에 비해 문법이 쉽고 간단하여 HTML 등의 서식 문서들로 쉽게 변환되기 때문에 README 파일이나 온라인 게시물 등에 자주 사용된다.

## 목차(header)
- # h1
- ## h2
- ### h3
- #### h4
- ##### h5
- ###### h6

## 리스트(list)
1. 장점
    - 간편하게 쉽다.
    - 다양한 플랫폼에서 지원한다. 
2. 단점
    * 표준이 아니다
    * 표현매체마다 다르게 보인다
    * 모든 **HTML**을 대체하지 못한다.

1. **내용**: 강조
2. *내용*: 기울이기
3. ~~내용~~: 취소선
4. 내<br />용: 줄바꿈

- <sup>내용</sup>: 위 첨자
- o(nlog<sup>2</sup>)

- <sub>내용</sub>: 아랫첨자
- 홍길동<sub>작사</sub>

- []: 대괄호(브레이킷)
- {}: 중괄호(컬리 브레이스)
- (): 소괄호(패런서시스)
