```python
# list의 모든 합을 구하기
# min(): 가장 작은 값을 찾아서 반환하는 함수
# max(): 가장 큰 값을 찾아서 반환하는 함수
# sum(): 모든 값을 더하는 함수

sum(li1)
min(li1)
max(li1)
```
