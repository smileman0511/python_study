## keywards argument

```python
def order_info(*args, name, address, phone):
    total = sum(args)
    print(f"{name}님의 주소: {address}, 연락처: {phone}")
    print(f"총 구매 가격: {total}")
```

```python
order_info(1000, 50000, 15000, 7000, **user_info)
```

```python
# 국어, 영어, 수학 점수 평균 구하기
# 자릿수는 2번째 자리부터 반올림
# 사용자의 정보가 입력되면 같이 출력하기
def get_average(**kwards):
    kor, eng, math = kwards.get("kor", 0), kwards.get("eng", 0), kwards.get("math", 0)
    total = kor + eng + math

    average = total / 3
    
    if "round" in kwards:
        average = round(average, kwards["round"])
    
    if "name" in kwards:
        name = kwards["name"]
        return f"{name}님의 평균은: {average}"

    return f"평균은: {average}"
```

```python
user_info = {
    "kor": 100,
    "eng": 80,
    "math": 22,
    "name": "홍길동",
    "round": 5
}
```

```python
get_average(**user_info)
```

```python
# 두 값의 사칙연산 중 계산하고 싶은 값만 리턴해주기
# calc가 있을 때에만 계산해서 리턴, 없을 때에는 (10, 20)을 리턴

# 1. 두 개의 변수는 따로 변수에 받기
# 2. 부호는 따로 keywards 문법으로 받기

# ex) 10, 20, "+"

def calc(num1, num2, **kwards):
    if "calc" not in kwards:
        return (num1, num2)
    
    sign = kwards.get("calc")
    
    if sign == "+":
        return num1 + num2
    elif sign == "-":
        return num1 - num2
    elif sign == "*":
        return num1 * num2
    elif sign == "/":
        return num1 / num2
```

```python
print(calc(10, 20))
print(calc(10, 20, calc="+"))
print(calc(10, 20, calc="-"))
print(calc(10, 20, calc="*"))
print(calc(10, 20, calc="/"))
```
