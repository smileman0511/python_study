```python
# 국어, 영어, 수학 점수 평균 구하기
# 자릿수는 2번째 자리부터 반올림
# 사용자의 정보가 입력되면 같이 출력하기
def get_average(**kwards):
    kor, eng, math = kwards.get("kor", 0), kwards.get("eng", 0), kwards.get("math", 0)
    total = kor + eng + math

    average = total / 3
    
    if "round" in kwards:
        average = round(average, kwards["round"])
    
    if "name" in kwards:
        name = kwards["name"]
        return f"{name}님의 평균은: {average}"

    return f"평균은: {average}"
```

```python
user_info = {
    "kor": 100,
    "eng": 80,
    "math": 22,
    "name": "홍길동",
    "round": 5
}
```

```python
get_average(**user_info)
```

```python
# 두 값의 사칙연산 중 계산하고 싶은 값만 리턴해주기
# calc가 있을 때에만 계산해서 리턴, 없을 때에는 (10, 20)을 리턴

# 1. 두 개의 변수는 따로 변수에 받기
# 2. 부호는 따로 keywards 문법으로 받기

# ex) 10, 20, "+"
def calculator(num1, num2, **keywards) -> int:

    calc = keywards.get("calc", None)
    calc_set = {"+", "-", "*", "/", "%"}
    is_confirm = calc in calc_set

    if not is_confirm:
        print("부호는 +, -, *, /, %만 사용할 수 있습니다.")
        return -1
    
    num1 = str(num1)
    num2 = str(num2)
    
    if calc:
        return eval(f"{num1}{calc}{num2}")

    return num1, num2
```

```python
calculator(10, 20, calc="**")
```

```python
# 글자의 개수를 찾아주는 함수
# ex) 
# "apple", "a" -> 1
# "apple", "p" -> 2

# 파이썬은 증감식이 없다. ++, --
# +=
def count_char(content, c):
    count = 0
    for i in content:
        if i == c:
            count += 1

    return count
```

```python
count_char("abcabc", "a")
```

```python
def count_chars(content):
    result = {}

    for c in content:
        if c not in result:
            result.update({c: content.count(c)})
        
    return result
```

```python
count_chars("apple")
```
