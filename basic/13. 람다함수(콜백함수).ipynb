## map, filter

```python
li1 = [(i + 1) * 10 for i in range(10)]
```

```python
li1
```

```python
def multiple_2(num):
    return num * 2

# map(함수, 객체)
list(map(multiple_2, li1))

# 람다함수는 값을 리턴
list(map(lambda i: i * 2, li1))
```

```python
# 65 ~ 90 (A~Z)
li2 = [i for i in range(65, 91)]

print(li2)
```

```python
li3 = list(map(lambda c: chr(c), li2))
print(li3)
```

```python
# li3 소문자로 변경 li4를 생성
li4 = list(map(lambda c: chr(ord(c) + 32), li3))
print(li4)
```

```python
# li2로 aBcDeFgHiJ...로 변경 후 li5를 생성
li5 = list(map(lambda tu: chr(tu[1] + 32) if tu[0] % 2 == 0 else chr(tu[1]), enumerate(li2)))
print(li5)

result = "".join(li5)
print(result)
```

```python
def extract_char(tu):
    idx, num = tu
    return chr(num + 32) if idx % 2 == 0 else chr(num)

li5 = list(map(extract_char, enumerate(li2)))
print(li5)
```

```python
li6 = [i + 1 for i in range(100)]
print(li6)
```

```python
# 3의 배수 
li7 = list(filter(lambda n: n % 3 == 0, li6))
print(li7)
```

```python
# li6에 3의 배수들만 걸러내고 2배로 값을 바꾸기
li8 = list(map(lambda n: n * 2, filter(lambda n: n % 3 == 0, li6)))
print(li8)
```

```python
# li4에서 a ~ g까지만 filter로 걸러내고 li9 만들기
# li9 = 
li9 = list(map(lambda tu: tu[1], filter(lambda tu: tu[0] < 7, enumerate(li4))))
print(li9)
```
