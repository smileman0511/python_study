```python
# 완전 객체 지향형 언어
# Market
# 판매
class Market:
    def __init__(self, market_name, product_name, product_price):
        self.market_name = market_name
        self.product_name = product_name
        self.product_price = product_price

    def sell(self, customer):
        customer.money -= self.product_price * (100 - customer.discount) / 100

# Customer
# 할인율
class Customer:
    def __init__(self, name, money, discount):
        self.name = name
        self.money = money
        self.discount = discount
```

```python
cu = Market("CU", "도시락", 10000)
customer = Customer("이규혁", 50000, 10)

cu.sell(customer)
print(customer.money)
```

```python
# Atm, User 객체
# deposit

# Atm.deposit(User, 10000)
# 유저의 돈은 10000 차감,
# 은행에 돈은 10000 입금

# 은행이 가진돈과 유저의 돈 출력
```

```python
class Atm:
    def __init__(self, name):
        self.name = name
        self.users = {}

    def deposit(self, user, money):

        if user.money < money: 
            return "잔액 부족"
        
        if user.name in self.users:
            # 기존의 돈
            self.users.update({user.name: self.users[user.name] + money})
            
        else:
            self.users.update({user.name: money})

        user.money -= money

    def show(self):
        print(self.users)
```

## 상품 CRUD

```python
class Store:
    def __init__(self, name, products = {}, income = 0):
        self.name = name
        self.products = products
        self.income = income
        
    # 상품 등록
    def register(self, **product):
        if "product_name" in product:
            product_name = product.pop("product_name")

            # 유효성 검사
            if product_name not in self.products:
                # 기존의 상품이 없다면 추가
                self.products.update({product_name: product})
                return 
                
            else:
                # 기존의 상품이 있다면 추가하지 않는다.
                return

    def show(self):
        print("판매 상품 목록")
        print(self.products)
                

    def sell(self, user, product_name, quantity):
        # 상품 존재 여부 확인
        if product_name not in self.products:
            print("존재하지 않는 상품입니다.")
            return

        product = self.products[product_name]

        # 재고 확인
        if product["product_stock"] < quantity:
            print("재고가 부족합니다.")
            return

        total_price = product["product_price"] * quantity

        # 사용자 돈 확인
        if user.money < total_price:
            print("잔액이 부족합니다.")
            return

        # 판매 처리
        product["product_stock"] -= quantity
        user.money -= total_price
        self.income += total_price

        print(f"{product_name} {quantity}개 판매 완료")
    
    # 상품 판매
    # 수익 store 저장
    # 상품의 stock 구매수량만큼 제거
    # 구매자의 money도 빠져야함

    # 상품 수정
    # 이름, 가격, 재고 수정
    def modify(self, product_name, **kwargs):
        # 상품 존재 여부 확인
        if product_name not in self.products:
            print("존재하지 않는 상품입니다.")
            return

        product = self.products[product_name]

        # 이름 수정
        if "product_name" in kwargs:
            new_name = kwargs["product_name"]

            # 중복 이름 방지
            if new_name in self.products:
                print("이미 존재하는 상품명입니다.")
                return

            self.products[new_name] = self.products.pop(product_name)
            product_name = new_name
            product = self.products[product_name]

        # 가격 수정
        if "product_price" in kwargs:
            product["product_price"] = kwargs["product_price"]

        # 재고 수정
        if "product_stock" in kwargs:
            product["product_stock"] = kwargs["product_stock"]

        print("상품 수정 완료")

    # 상품 삭제
    def remove(self, product_name):
        # 상품 존재 여부 확인
        if product_name not in self.products:
            print("존재하지 않는 상품입니다.")
            return

        del self.products[product_name]
        print("상품 삭제 완료")

class User:
    def __init__(self, name, money):
        self.name = name
        self.money = money
```

```python
gs_store = Store("GS")

product1 = {
    "product_name": "노트북",
    "product_price": 100_000,
    "product_stock": 50
}

product2 = {
    "product_name": "노트북",
    "product_price": 100_000,
    "product_stock": 50
}

gs_store.register(**product1)
gs_store.register(**product2)
gs_store.show()

user1 = User("철수", 500_000)

# 판매
gs_store.sell(user1, "노트북", 2)

# 수정
gs_store.modify(
    "노트북",
    product_name="게이밍노트북",
    product_price=200_000,
    product_stock=30
)

# 삭제
gs_store.remove("게이밍노트북")

gs_store.show()

print("상점 수익:", gs_store.income)
print("유저 잔액:", user1.money)
```

```python
# 클래스 메서드
class Korean:
    # 클래스 변수
    country = "한국"

    # 클래스 메서드
    # 인스턴스X, 클래스 변수 O
    # 데코레이터
    @classmethod
    def trip(cls, country):
        if cls.country == country:
            print("국내여행")
        else:
            print("해외여행")

    # 정적 메서드
    @staticmethod
    def slogan():
        print("Imagine your korea")
```

```python
hong = Korean()

hong.trip("한국")
hong.slogan()

Korean.trip("미국")
Korean.slogan()
```

# 상속(Inheritance)

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age
```

```python
hong = Person("홍길동")
jang = Student("장보고", 20)

# 부모 -> 타입도 같이 전달!
# 모든 자식은 부모 타입이다.
# 부모는 자식타입이 아니다.
print(isinstance(hong, Person))
print(isinstance(hong, Student))
print(isinstance(jang, Person))
print(isinstance(jang, Student))
```

```python
class Animal:
    def __init__(self, name, feed, species):
        self.name = name
        self.feed = feed
        self.species = species

    def feeding(self):
        print(f"{self.name}은 먹이를 먹습니다.")
        

class Dog(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)
    
    def act(self):
        print(f"{self.name}은 산책을 합니다.")

class Cat(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)

    def sleep(self):
        print(f"{self.name}은 잠을 자주 잡니다.")
```

```python
dog = Dog("누렁이", "잡식", "강아지")
dog.act()
dog.feeding()

cat = Cat("야옹이", "츄르", "고양이")
cat.sleep()
cat.feeding()
```

```python
class Person:
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    def hello(self):
        print(f"{self.name}님이 인사를 합니다.")

class Student(Person):
    # 1. 생성자에서 phone 매개변수를 추가로 받아야 합니다.
    def __init__(self, name, age, height, weight, phone):
        # 2. super()에는 부모 클래스의 생성자가 기대하는 인자만 전달합니다.
        super().__init__(name, age, height, weight)
        # 3. 자식 클래스만의 고유 속성인 phone을 저장합니다.
        self.phone = phone

    def study(self):
        print(f"{self.name}님은 {self.phone} 번호의 핸드폰을 끄고 공부를 합니다.")

# 4. 객체 생성
student1 = Student("홍길동", 26, 178, 78, "010-1234-5678")

student1.hello() # 부모의 메서드 호출
student1.study() # 자식의 메서드 호출
```

```python
class Car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def start_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 켭니다.")

    def stop_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 끕니다.")

class SuperCar(Car):
    def start_engine(self):
        print(f"{self.brand}는 음성으로 시동을 켭니다.")

    def stop_engine(self):
        print(f"{self.brand}는 음성로 시동을 끕니다.")
```

```python
matiz = Car("마티즈", 10000000)
matiz.start_engine()
matiz.stop_engine()
```

```python
lambo = SuperCar("람보", 50000000)
lambo.start_engine()
lambo.stop_engine()
```
