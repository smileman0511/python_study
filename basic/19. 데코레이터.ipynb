```python
from datetime import datetime

def add_number(num1, num2):
    print(datetime.now())
    return num1 + num2
```

```python
add_number(10, 20)
```

```python
def concat(*args):
    print(datetime.now())
    return "".join(map(lambda data: str(data), args))
```

```python
concat(10, 20, 30)
```

```python
def log_time(original_function):
    def logging(*args):
        print(datetime.now())
        return original_function(*args)
    return logging
```

```python
add_number(10, 20)
```

```python
@log_time
def add_number(num1, num2):
    return num1 + num2
```

```python
add_number(10, 20)
```

```python
def divide(num1, num2):
    try:
        return int(num1) / int(num2)
    except ZeroDivisionError as e:
        print("0으로 나눌 수 없습니다")
    except ValueError as e:
        print("정수만 입력할 수 있습니다.")
    except Exception as e:
        print("알 수 없는 오류가 발생했습니다.")
```

```python
def subtract(num1, num2):
    try:
        return num1 - num2
    except TypeError as e:
        print("정수만 계산할 수 있습니다.")
```

```python
divide(10, "ㄱ")
```

```python
subtract("ㄱ", 10)
```

```python
def safe_calc(original_function):
    def exception_wrapper(*args):
        try:
            return original_function(*args)
        except ZeroDivisionError as e:
            print("0으로 나눌 수 없습니다")
        except ValueError as e:
            print("정수만 입력할 수 있습니다.")
        except TypeError as e:
            print("정수만 계산할 수 있습니다.")
        except Exception as e:
            print("알 수 없는 오류가 발생했습니다.")
            
    return exception_wrapper
```

```python
@safe_calc
def divide(num1, num2):
        return num1 / num2
```

```python
@safe_calc
def subtract(num1, num2):
    return num1 - num2
```

```python
divide(10, 0)
```

```python
subtract("ㄱ", 1)
```

```python
# 이름과 정수를 몇 개 입력할 지 모를 때
# "홍길동님이 전달한 총 합: 000"으로 출력해주는 메서드 구현
```

```python
# KeyError, TypeError, ValueError                               
def safe_value(original_function):
    def exception_wrapper(*args, **keywards):
        try:
            return original_function(*args, **keywards)
        except KeyError:
            print("이름을 전달하세요.\nex)name=\'홍길동\'")
        except TypeError:
            print("정수만 더할 수 있습니다.")
        except Exception as e:
            print(e)
    return exception_wrapper


@safe_value
def print_total(*args, **kwards):
    name = kwards["name"]
    total = sum(args)
    result = f"{name}님이 전달한 총 합: {total}"
    print(result)
```

```python
print_total(10, 20, 30, 40, name="홍길동")
```
