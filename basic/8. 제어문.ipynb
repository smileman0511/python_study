```python
score = int(input("점수를 입력하세요.ex)60\n"))
```

```python
grade = ''

if score > 90:
    grade = 'A'
elif score > 80:
    grade = 'B'
elif score > 70:
    grade = 'C'
else:
    grade = 'F'

print(grade)
```
