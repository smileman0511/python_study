## 컴프리헨션

```python
[i * 2 for i in [10, 20, 30]]
```

```python
{i * 2 for i in [10, 20, 30]}
```

```python
name = "장보고", "홍길동", "이순신"
```

```python
for i in name:
    print(i)
```

```python
# enumerate: 인덱스를 만드는 함수
# [(0, 장보고), (1, 홍길동), (2, 이순신)]
# unpacking
for index, name in enumerate(name):
    print(f"{index + 1}번 {name}")
```

```python
# 일반 for문(range)
# range(10) -> ~ 9
# range(10, 20) -> 10 ~ 19
# range(20, 30, 2) -> 20 ~ 29, 2개씩

for i in range(10):
    print(i)
```

```python
# 1 ~ 50까지 홀수만 출력하기
for i in range(1, 50, 2):
    print(i)
```

```python
# 1 ~ 50까지 홀수만 리스트에 담아서 출력
# 컴프리헨션 문법 사용
odd_list = [i for i in range(1, 50, 2)]
print(odd_list)
```

```python
# 1 ~ 50까지 반복하여
# 세트 컴프리헨션에 0,1,2,3,4,5만 담아서 출력
set1 = {i % 6 for i in range(50)}

print(set1)
```

## 딕셔너리 빠른 for문

```python
foods = {
    "a": "카레",
    "b": "라면",
    "c": "불고기",
    "d": "떡볶이",
}
```

```python
# index 접근
for i in foods:
    print(foods.get(i))
```

```python
for i in foods.values():
    print(i)
```

```python
for i in list(foods.values()):
    print(i)
```

```python
for key, food in foods.items():
    print(key, food)
```
